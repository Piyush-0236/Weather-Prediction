```python
#IMPORTING LIBRARIES
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np

#IMPORTING DATASET
df = pd.read_csv('weather_classification_data.csv')
df
```

```python
# DROPPING DUPLICATE VALUES
df.drop_duplicates(inplace = True)
```

```python
# FINDING UNIQUE VALUES
df['Cloud Cover'].unique()
```

```python
# FINDING UNIQUE VALUES
df['Location'].unique()
```

```python
# MAPPING STRING VALUES TO INTEGER
df['Cloud Cover'] = df['Cloud Cover'].map({'partly cloudy':0,'clear':1,'overcast':2,'cloudy':3})
df['Season'] = df['Season'].map({'Winter':0,'Spring':1,'Summer':2,'Autumn':3})
df['Location'] = df['Location'].map({'inland':0,'mountain':1,'coastal':2})
```

```python
# MAPPING STRING VALUES TO INTEGER
df['Weather Type'] = df['Weather Type'].map({'Rainy':0,'Cloudy':1,'Sunny':2,'Snowy':3})
df
```

```python
# BOXPLOT FOR EACH COLUMN
plt.figure(figsize = (10,5))
sb.boxplot(data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
# REMOVING OUTLIERS
for i in df.columns:
  q1 = df[i].quantile(0.25)
  q3 = df[i].quantile(0.75)
  iqr = q3 - q1
  upper = q3+1.5*iqr
  lower = q1-1.5*iqr
  df = df[(df[i]>=lower) & (df[i]<=upper)]
```

### LOGISTIC REGRESSION

```python
# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
# BUILDING THE LOGISTIC REGRESSION MODEL
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
# PREDICTION BY THE MODEL
y_pred = lr.predict(x_test)
y_pred
```

```python
# DETERMINING ACCURACY OF THE MODEL
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a = confusion_matrix(y_test,y_pred)
print("Accuracy Score : ", accuracy_score(y_test,y_pred)*100)
print(classification_report(y_test,y_pred))
print(sb.heatmap(a,annot=True))
```
